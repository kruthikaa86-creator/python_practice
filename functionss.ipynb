```python
def comples_function(a, b, *args, c = 10, **kwargs):
    print(f"a (position): {a}")
    print(f"b (position): {b}")
    print(f"args (variable position): {args}")
    print(f"c (default): {c}")
    print(f"*kwargs (variable keyword): {kwargs}")

comples_function(1,2,3,4, c = 20, x = 100)
```

```python
def calculate_stats(numbers):
    minimum = min(numbers)
    maximun = max(numbers)
    avrage = sum(numbers)/len(numbers)

    return minimum,maximun,avrage

min_val, max_val, avg_val = calculate_stats([1,2,3,4,2,3,4,8])
```

```python
def print_report(data):
    if not data:
        print("no data to display")
        return 
    print("data report: ")
    for item in data:
        print(f"{item}")

print_report([])
```

```python
nums = [12,34,54,45,34,45,34,54,98]
def cal_mean(nums):
    return sum(nums)/len(nums)

def cal_median(nums):
    sorted_nums = sorted(nums)
    n = len(sorted_nums)

    if n % 2 == 0:
        mid1 = sorted_nums[n // 2 - 1]
        mid2 = sorted_nums[n//2]
        return (mid1 + mid2) / 2

    else:
        return sorted_nums[n//2]
    pass
def cal_mode(nums):
    freq = {}
    for num in nums:
        if num in freq:
            freq[num] = freq[num] + 1
        else:
            freq[num] = 1
    max_count = max(freq.values())
    mode = []

    for num in freq:
        if freq[num] == max_count:
            mode.append(num)
    return mode
print(cal_mean(nums))
print(cal_median(nums))
print(cal_mode(nums))
```
