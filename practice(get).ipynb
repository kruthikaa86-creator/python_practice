```python
nums = [1,2,2,3,3,4]
freq = {}
for n in nums:
    freq[n] = freq.get(n,0) + 1 # returns 0 if n is not there later adds 1
print(freq)
```

```python
s = 'interview'
freq = {}
for ch in s:
    freq[ch] = freq.get(ch,0)+1
print(freq)
```
