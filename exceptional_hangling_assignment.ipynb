```python
#Q1.
#Write a Python program that takes two integers from the user and prints their division.
#Handle the case where the second number is zero.

try:
    num1 = int(input("enter the number 1:"))
    num2 = int(input("enter the number 2:"))

    div = num1/num2
    print("division result:",div)

except ZeroDivisionError:
    print("error: cannot divide by zero")
except ValueError:
    print("error: please enter valid integers")
```

```python
# Q2. Write a Python program that:
# Creates a list of 5 integers
# Asks the user for an index
# Prints the element at that index
# Handles the case when the index is out of range

try:
    list1 = [23,54,65,67,93]
    index = int(input("enter the index: "))
    print("Element at index:", list1[index])

except IndexError:
    print("Error: Index out of range")
except ValueError:
    print("Error: Please enter a valid integer")
```

```python
try:
    num1 = int(input("enter the number 1:"))
    num2 = int(input("enter the number 2:"))
    div = num1 / num2
    print("division result:",div)

except ZeroDivisionError:
    print("error: enter non zero integer")
    
except ValueError:
    print("enter valid integer")

finally:
    print("program finished execution") 
```

```python
# only positive number
try:
    num = int(input("enter the number:"))

    if num < 0:
        raise ValueError("negative numbers are not allowed")
    print("valid number:",num)

except ValueError as e :
    print("error:",e)
```

```python
try:
    age = int(input("enter your age:"))
    
    if age < 18:
        raise ValueError("age is below 18")
    print("enter valid age:",age)
    if age >= 18:
        print("access granted")

except ValueError as e:
    print("error:",e)
```

```python
# Write a function check_age(age)

def check_age(age):
    if age < 18:
        raise ValueError("age is below 18")
    print("access granted")

try:
    age = int(input("enter your age:"))
    check_age(age)

except ValueError as e:
    print("error:",e)
```
