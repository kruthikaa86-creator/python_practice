```python
add = lambda a,b : a + b
add(12,4)
```

```python
square = lambda x : x**2
square(5)
```

```python
area = lambda length, width : length * width
area(4,4)
```

```python
is_even = lambda x: "even" if x % 2 == 0 else "odd"
is_even(8)
```

```python
reversed_str = lambda s:s[::-1]
reversed_str("hey")
```

```python
celcius = [0,10,30,40,20]
fahrenhiet = list(map(lambda c : (c * 9/5) + 32,celcius)) #map = lamda funct on each element
fahrenhiet
```

```python
name = "kruthika","nitin"

upper_name = list(map(lambda name : name.upper(),name))
print(upper_name)
```

```python
num = [1,2,3,4,5,6]
result = list(filter(lambda x: x%2 == 0,num)) #filter -> returns value which is true for the condition

result
```

```python
students = [
    {"name": "A", "marks": 87},
    {"name": "B", "marks": 60},
    {"name": "C", "marks": 56}
]
results = list(filter(lambda x : x['marks']<60,students))
results
```
